```python
from langchain_core.runnables import RunnableParallel
from langchain_core.exceptions import OutputParserException
import logging
```

```python
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
```

```python
retriever = vector_store.as_retriever(
    search_type="similarity_score_threshold",
    search_kwargs={"score_threshold": 0.7, "k": 4}
)
```

```python
refined_prompt = PromptTemplate(
    template="""You are an assistant for question-answering tasks. Use only the provided context to answer the question. If the context does not contain enough information to answer, say "I don't have enough information to answer this question." Keep the answer concise, using up to three sentences.
Question: {question}
Context: {context}
Answer:""",
    input_variables=["context", "question"]
)
```

```python
def format_docs(docs):
    if not docs:
        return "No relevant documents found."
    return "\n\n".join(f"Source: {doc.metadata['source']}\nContent: {doc.page_content}" for doc in docs)
```

```python
rag_chain = (
    RunnableParallel({"context": retriever | format_docs, "question": RunnablePassthrough()})
    | refined_prompt
    | model
    | StrOutputParser()
)
```

```python
test_queries = [
    "What is LangChain used for?",
    "What did someone have for breakfast?",
    "What is the weather forecast?",
    "What is Llama model?",
    "Who are the top soccer players?"
]

print("Testing RAG System with Multiple Queries:\n")
for query in test_queries:
    try:
        answer = rag_chain.invoke(query)
        print(f"Question: {query}")
        print(f"Answer: {answer}\n")
    except Exception as e:
        logger.error(f"Error processing query '{query}': {str(e)}")
        print(f"Question: {query}")
        print(f"Answer: An error occurred: {str(e)}\n")
```
